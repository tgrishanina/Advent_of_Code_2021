Day 1 Part 1

```python
file = open('input.txt', 'r')
row_lists = []
for row in file.readlines():
    row = row.rstrip("\n")
    row_lists.append(int(row))
```

```python
print(row_lists)
```

```python
increase_list = []
for i in range(len(row_lists) - 1):
    if row_lists[i+1] > row_lists[i]:
        increase_list.append(row_lists[i+1])

print(len(increase_list))
```

```python
sums_list = []
for i in range(len(row_lists) - 2):
    window_sum = row_lists[i] + row_lists[i+1] + row_lists[i+2]
    sums_list.append(window_sum)
sums_increase_list = []
for i in range(len(sums_list) - 1):
    if sums_list[i+1] > sums_list[i]:
        sums_increase_list.append(sums_list[i+1])

print(len(sums_increase_list))
```

Day 2 Part 1

```python
file = open('input_d2.txt', 'r')
row_list2 = []
for row in file.readlines():
    row = row.rstrip("\n")
    row_list2.append(row)
```

```python
d = 0
h = 0
for row in row_list2:
    word, num = row.split()
    num = int(num)
    if word == "forward":
        h = h + num
    if word == "down":
        d = d + num
    elif word == "up":
        d = d - num
print(d)
print(h)
print(d*h)
```

Day 2 Part 2

```python
d = 0
h = 0
a = 0
for row in row_list2:
    word, num = row.split()
    num = int(num)
    if word == "forward":
        h = h + num
        d = d + (a*num)
    elif word == "down":
        a = a + num
    elif word == "up":
        a = a - num
print(d*h)
```

Day 3 Part 1

```python
file = open('input_d3.txt', 'r')
row_list3 = []
for row in file.readlines():
    row = row.rstrip("\n")
    row_list3.append(row)
```

```python
from statistics import mode

gamma = ""
for i in range(len(row_list3[0])):
    nums_list = []
    for row in row_list3:
        nums_list.append(row[i])
    gamma = gamma + mode(nums_list)
    
print(gamma)      
```

```python
epsilon = "101011000110"
g = int(gamma, 2)
e = int(epsilon, 2)
print(g)
print(e)
```

Day 3 Part 2

```python
from collections import Counter

temp_list = row_list3[:]  # Start with a copy to keep the original list unchanged

while len(temp_list) > 1: 

    # Loop through columns (indexes of row elements)
    for i in range(len(temp_list[0])):
        nums_list = [row[i] for row in temp_list]  # Get all values in the `i`-th column

        vals = Counter(nums_list)

        if len(set(vals.values())) == 1:
            most_common = "1"
        else:
            most_common = max(vals, key=vals.get)
        
        # Filter temp_list to include only rows that match the current column's mode
        temp_list = [row for row in temp_list if row[i] == most_common]
        
        # Debugging output to observe the changes
        print(nums_list)
        print(f"Column {i}, mode: {most_common}")
        print(f"Filtered temp_list: {temp_list}")
        
        # Exit if only one row remains
        if len(temp_list) == 1:
            break

print("Final remaining rows in temp_list:", temp_list)
```
